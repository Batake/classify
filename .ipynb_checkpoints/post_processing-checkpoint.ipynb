```python
from functools import reduce
import glob
import numpy as np
import pandas as pd

class MakeLabelFile:
    def __init__(self, data_per_second, st_threshold, boundary, sec_start, sec_end, time_interval, convolve_range, observe_time):
        self.data_per_second = data_per_second
        self.st_threshold = st_threshold
        self.boundary = boundary
        self.sec_start = sec_start#秒後から見る
        self.sec_end = sec_end #秒前まで見る
        self.starts, self.ends, self.numbers, self.actions, self.states = [], [], [], [], []
        self.time_interval = time_interval
        self.convolve_box = np.ones(convolve_range)/convolve_range
        self.observe_time = observe_time


    def __judge_label(self, std_start, std_end):
        if std_start < self.label_start:
            start_detected = self.label_start
        elif self.label_start <= std_start and std_start <= self.label_end:
            start_detected = std_start
        else:
            start_detected = self.label_end

        if std_end < self.label_start:
            end_detected = self.label_start
        elif self.label_start <= std_end and std_end <= self.label_end:
            end_detected = std_end
        else:
            end_detected = self.label_end    

        if (end_detected - start_detected)/(self.label_end - self.label_start) >= self.boundary:
            label = FileOperation.label_from_action(self.action)#actionにより1~7のlabelを返す
        else:
            label = [8]

        return label


    def __extend_datas(self):
        start_flag, end_flag = 0, 0
        called_flag = 0
        data_interval = int(self.data_per_second * self.time_interval)
        for data_i in range(self.data_per_second*self.sec_start, len(self.distance) - self.data_per_second * self.sec_end, data_interval):
            if np.std(self.distance[data_i:data_i+data_interval]) >self.st_threshold:
                if start_flag == 0:
                    start_flag = 1
                    self.starts.extend([data_i/self.data_per_second])
                    self.numbers.extend([self.file_index])
                    self.actions.append(self.action)
                    if (called_flag == 1) and (self.starts[-1] - self.ends[-1] < self.observe_time):
                        del self.starts[-2], self.numbers[-2], self.actions[-2], self.ends[-1], self.states[-1]
                    
            else:
                if start_flag == 1:
                    start_flag = 0
                    called_flag = 1
                    self.ends.extend([data_i/self.data_per_second])
                    label = self.__judge_label(std_start = self.starts[-1], std_end = self.ends[-1])
                    self.states.append(label)

        if len(self.ends) != len(self.starts):
            self.ends.extend([data_i/self.data_per_second])
            label = self.__judge_label(std_start = self.starts[-1], std_end = self.ends[-1])
            self.states.append(label)
        
        
    def make_label_file(self, data_folder, time_file, exclude_actions, save_file):
        files = []
        files = sorted(glob.glob(data_folder))
        source = pd.read_csv(time_file, delimiter=',', encoding='cp932', engine='python')
        files_excluded = FileOperation.set_file_excluded(files, exclude_actions)
        source_excluded = FileOperation.set_source_excluded(source, exclude_actions)
        labels_start, labels_end = source_excluded["change_start_s"], source_excluded["change_end_s"]
        
        for file_index, file in enumerate(files_excluded):
            self.file_index = file_index
            self.action = FileOperation.get_action(file)[:-1]
            self.distance_original = pd.read_csv(file, skiprows= 0, header=None, names=['distance'], engine='python')['distance']
            self.distance = np.convolve(self.distance_original, self.convolve_box, 'same')
            self.label_start, self.label_end = labels_start[file_index], labels_end[file_index]
            self.__extend_datas()
            
#         all_action = reduce(lambda x,y: x+y, self.actions)
        data_extended = pd.DataFrame({"number":self.numbers, "change_start_s":self.starts, "change_end_s":self.ends, "state":self.states, "act": self.actions})
        data_extended.to_csv(save_file)
```

```python
def search_time(distance, st, details, change_start_s, change_end_s, k):    
    see_time = 1.0 #変動がないかを見る時間(s)の範囲
    label_time = see_time #１ラベルあたりの時間
    feature_bin = int(label_time * Sa / window)
    window_num = int(see_time  / label_time)#遡って見るまとまりの数 ここでは10個

#     print("標準偏差")
    i = 27
    start, end, number, action, state = [], [], [], [], []
    start_flag, end_flag = 0, 0
    la = [0]
    left, right = 0, 0
    end_candidate = 0
    boundary = 0.8
    
    action_name = details
#     print(action_name)
    while (i<(len(distance)-54)):
        if np.std(distance[i:i+27]) > st:
            if start_flag == 0:
                start.extend([i/27])
                number.extend([k])
#                 print("start:" + str(i / 27))
                start_flag = 1
        else:
            if start_flag == 1:
                end.extend([i/27])
                start_flag = 0
#                 left = 0
#                 right = 0
                
                if start[-1] < change_start_s[k]:
                    left = change_start_s[k]
                elif change_start_s[k] <= start[-1] and start[-1] <= change_end_s[k]:
                    left = start[-1]
                else:
                    left = change_end_s[k]
                    
                if end[-1] < change_start_s[k]:
                    right = change_start_s[k]
                elif change_start_s[k] <= end[-1] and end[-1] <= change_end_s[k]:
                    right = end[-1]
                else:
                    right = change_end_s[k]
                
#                 print(left)
#                 print(right)
#                 if (right - left)/(change_end_s[j] - change_start_s[j]) == 0:
#                     print(j)
#                     print((right - left)/(change_end_s[j] - change_start_s[j]))
                
                if (right - left)/(change_end_s[k] - change_start_s[k])>=boundary:
                    if action_name == "fall" or action_name == "standfall":
                        la = [1]
                    elif action_name == "front":
                        la = [2]
                    elif action_name == "side":
                        la = [3]
                    elif action_name == "back":
                        la = [4]
                    elif action_name == "bodychange" or action_name == "other":
                        la = [5]
                    elif action_name == "sit":
                        la = [6]
                    elif action_name == "stand":
                        la = [7]
                    else:
                        la = [0]
                else:
                    la = [8]
                state.extend(la)
                action.append(action_name)
#                 print("end:" + str([i/27]))
        i = i + 27
        
        
    if len(start) != len(end):
        end.extend([i/27])
    if len(start) != len(state):
        if end[-1] < change_start_s[k]:
            right = change_start_s[k]
        elif change_start_s[k] <= end[-1] and end[-1] <= change_end_s[k]:
            right = end[-1]
        else:
            right = change_end_s[k]

        if (right - left)/(change_end_s[k]-change_start_s[k])>=boundary:
            if action_name == "fall" or action_name == "standfall":
                la = [1]
            elif action_name == "front":
                la = [2]
            elif action_name == "side":
                la = [3]
            elif action_name == "back":
                la = [4]
            elif action_name == "bodychange" or action_name == "other":
                la = [5]
            elif action_name == "sit":
                la = [6]
            elif action_name == "stand":
                la = [7]
            else:
                la = [0]
        else:
            la = [8]
            
        
        state.extend(la)
        action.append(action_name)
    return number, start, end, state, action


def find_label(st):
    data_path = '/Users/takabatakewataru/anaconda3/a_workspace/experiment/data/toilet/jrc_new/labeled_data/labeled_data_proposed/*_*.csv'
    files = []
    files = sorted(glob.glob(data_path))
    source = pd.read_csv('/Users/takabatakewataru/anaconda3/a_workspace/experiment/data/toilet/jrc_new/time_label/20180221.csv', delimiter=',', encoding='cp932')
    exclude_actions = ['no']
    files_excluded = set_file_excluded(files, exclude_actions)
    source_excluded = set_source_excluded(source, exclude_actions)
    change_start_s = source_excluded["change_start_s"]
    change_end_s = source_excluded["change_end_s"]
    before_state = source_excluded["before_state"]
    after_state = source_excluded["after_state"]


    # stds = np.linspace(0, 0.1, 11)
#     stds = [0.06]
    stds = [st]

    for q, st in enumerate(stds):
#         if st != 0.06:
#             continue
#         print(st)
        time_information = []
        dif = []
        dif2 = []
        start = []
        end = []
        number = []
        labels = []
        action = []
        global k
        for k, file in enumerate(files_excluded):
            details = get_file_details(file)["action"][:-1]
            distance = pd.read_csv(str(file), skiprows= 0, header=None, names=['real', 'distance','estimate'], engine='python')['distance']
            n, s, e, l, a = search_time(distance, st, details, change_start_s, change_end_s, k)

            number.extend(n)
            start.extend(s)
            end.extend(e)
            labels.extend(l)
            action.append(a)

        from functools import reduce
        action = reduce(lambda x,y: x+y,action)
        print(action)
        data = pd.DataFrame({"number":number, "change_start_s":start, "change_end_s":end, "state":labels, "act": action})
        data.to_csv("/Users/takabatakewataru/anaconda3/a_workspace/experiment/data/toilet/jrc_new/time_label/time.csv")
```
